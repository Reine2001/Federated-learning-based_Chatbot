```python
import PyPDF2

def extract_text_pypdf2(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text()
    return text

# Chemin relatif vers le fichier PDF
file_path = 'donnees/Autres/LOI N013-2019-AN.pdf'

# Test de l'extraction
text_pypdf2 = extract_text_pypdf2(file_path)
print("PyPDF2 Extraction:\n", text_pypdf2[:2000], "\n")
```

```python
import fitz  # PyMuPDF

def extract_text_pymupdf(file_path):
    doc = fitz.open(file_path)
    text = ''
    for page in doc:
        text += page.get_text()
    return text

# Test de l'extraction
text_pymupdf = extract_text_pymupdf(file_path)
print("PyMuPDF Extraction:\n", text_pymupdf[:2000], "\n")
```

```python
from PIL import Image
import pytesseract
import fitz  # PyMuPDF pour extraire les images des pages du PDF

def extract_text_ocr(file_path):
    doc = fitz.open(file_path)
    text = ''
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        pix = page.get_pixmap()
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
        text += pytesseract.image_to_string(img)
    return text
file_path = 'donnees/Police_Administrative/Dcret n 2005-045.pdf'
# Test de l'extraction
text_ocr = extract_text_ocr(file_path)
print("OCR Extraction:\n", text_ocr[:2000], "\n")
```

```python
from pdfminer.high_level import extract_text
import re
import json

# Chemin relatif vers le fichier PDF
file_path = 'donnees/Autres/LOI N013-2019-AN.pdf'

# Étape 1 : Extraction des données avec pdfminer.six (à partir de la troisième page)
def extract_text_pdfminer(file_path):
    from pdfminer.layout import LAParams
    from pdfminer.high_level import extract_pages
    text = ""
    for i, page_layout in enumerate(extract_pages(file_path, laparams=LAParams())):
        if i >= 2:  # Ignore les deux premières pages
            for element in page_layout:
                if hasattr(element, "get_text"):
                    text += element.get_text()
    return text

# Étape 2 : Nettoyage des données
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # Remplacer les espaces multiples par un seul espace
    text = re.sub(r'\d+\s+', '', text)  # Suppression des numéros de page
    return text

# Étape 3 : Structuration des données
def structure_text(text):
    # Ajustement de l'expression régulière pour capturer les articles
    articles = re.split(r'(Article \d+-)', text)
    
    # Vérification si la séparation des articles a bien fonctionné
    for i, article in enumerate(articles[:10]):
        print(f"Article part {i}: {article[:100]}")
    
    articles_dict = {}
    for i in range(1, len(articles), 2):
        article_title = articles[i]
        article_content = articles[i + 1].strip()
        articles_dict[article_title] = article_content
    
    return articles_dict

# Extraction du texte
text_pdfminer = extract_text_pdfminer(file_path)
print("pdfminer.six Extraction:\n", text_pdfminer[:2000], "\n")

# Nettoyage du texte
cleaned_text = clean_text(text_pdfminer)
print("Texte nettoyé:\n", cleaned_text[:2000], "\n")

# Structuration du texte
structured_data = structure_text(cleaned_text)

# Affichage des premiers articles pour vérifier
for article, content in list(structured_data.items())[:5]:
    print(article)
    print(content[:500])
    print()

# Sauvegarde des articles structurés en JSON
with open('donnees/Autres/articles.json', 'w') as json_file:
    json.dump(structured_data, json_file, ensure_ascii=False, indent=4)
```
