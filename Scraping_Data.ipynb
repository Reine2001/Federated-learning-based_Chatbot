```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin
import time

URL = "https://academiedepolice.bf/index.php/services-en-ligne/telechargement/category/14-securite"

# En-tête User-Agent pour faire passer la requête pour un navigateur
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

# Fonction pour effectuer la requête avec des tentatives de reconnexion
def get_with_retries(url, max_retries=5):
    retries = 0
    while retries < max_retries:
        try:
            response = requests.get(url, headers=headers, timeout=10)
            response.raise_for_status()  # Vérifie si la requête a réussi
            return response
        except requests.exceptions.RequestException as e:
            print(f"Erreur lors de la connexion : {e}")
            retries += 1
            time.sleep(2)  # Attendre avant de réessayer
    return None

# Effectuer la requête initiale
response = get_with_retries(URL)
if response is None:
    print("Impossible de se connecter au site après plusieurs tentatives.")
else:
    soup = BeautifulSoup(response.content, 'html5lib')

    # Créer un dossier pour stocker les fichiers PDF téléchargés
    pdf_folder = "Securité"
    if not os.path.exists(pdf_folder):
        os.makedirs(pdf_folder)

    # Trouver tous les liens vers les fichiers PDF
    for link in soup.find_all('a', href=True):
        href = link['href']
        if href.endswith('.pdf'):
            pdf_url = urljoin(URL, href)  # Utiliser urljoin pour gérer les liens relatifs
            
            try:
                # Télécharger le fichier PDF
                pdf_response = get_with_retries(pdf_url)
                if pdf_response is None:
                    print(f"Impossible de télécharger {pdf_url} après plusieurs tentatives.")
                    continue
                
                # Nom du fichier PDF
                pdf_name = os.path.join(pdf_folder, pdf_url.split('/')[-1])
                
                # Enregistrer le fichier PDF
                with open(pdf_name, 'wb') as pdf_file:
                    pdf_file.write(pdf_response.content)
                
                print(f"Téléchargé: {pdf_name}")
            
            except requests.exceptions.RequestException as e:
                print(f"Erreur lors du téléchargement de {pdf_url}: {e}")

    print("Tous les fichiers PDF téléchargeables ont été traités.")
```
